```python
#importing OpenCV library
import cv2
from matplotlib import pyplot as plt
```

```python
#Reading the image
image=cv2.imread(r"C:\Users\Ayush\Desktop\free-photo-of-great-grey-shrike-sitting-on-bush-with-leaves.jpeg")
```

```python
#image dimensions
image.shape
```

```python
#displaying image using cv
cv2.imshow("image",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#displaying image using matplotlib
plt.imshow(image)
```

The default format here is BGR so get the original image we need to convert it to RGB

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

Now Converting it to Grayscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("New Dog", gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(gray_image)
```

invert the new grayscale image

```python
inverted_image = 255 - gray_image
cv2.imshow("Inverted", inverted_image)
cv2.waitKey()
```

blur the image by using the Gaussian Function in OpenCV

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```
